```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
train=pd.read_csv('/content/drive/MyDrive/DataSet/train (1).csv')
train.head(1)
train.shape
train[train['Embarked'].isnull()]
```

```python
Embarked=train.Embarked.value_counts()
Embarked
```

```python
S_Embarked=train[train['Embarked']=="S"]
S_Embarked_counts=len(S_Embarked)
```

```python
C_Embarked=train[train['Embarked']=="C"]
C_Embarked_counts=len(C_Embarked)
```

```python
Q_Embarked=train[train['Embarked']=="Q"]
Q_Embarked_counts=len(Q_Embarked)
```

```python
plt.pie([S_Embarked_counts,C_Embarked_counts,Q_Embarked_counts],labels=['S_Embarked_counts','C_Embarked_counts','Q_Embarked_counts'],autopct='%1.1f%%')
plt.title("Embarked wise passengers in Titanic")
plt.show()
```

```python
S_Embarked_survived=S_Embarked[S_Embarked['Survived']==1]
S_Embarked_unsurvived=S_Embarked[S_Embarked['Survived']==0]
Q_Embarked_survived=Q_Embarked[Q_Embarked['Survived']==1]
Q_Embarked_unsurvived=Q_Embarked[Q_Embarked['Survived']==0]
C_Embarked_survived=C_Embarked[C_Embarked['Survived']==1]
C_Embarked_unsurvived=C_Embarked[C_Embarked['Survived']==0]


S_embarked_survival_pct = len(S_Embarked_survived) / len(S_Embarked) * 100
S_embarked_unsurvival_pct = len(S_Embarked_unsurvived) / len(S_Embarked) * 100
Q_embarked_survival_pct = len(Q_Embarked_survived) / len(Q_Embarked) * 100
Q_embarked_unsurvival_pct = len(Q_Embarked_unsurvived) / len(Q_Embarked) * 100
C_embarked_survival_pct = len(C_Embarked_survived) / len(C_Embarked) * 100
C_embarked_unsurvival_pct = len(C_Embarked_unsurvived) / len(C_Embarked) * 100


fig, ax = plt.subplots()
ax.bar(['S', 'Q', 'C'], [S_embarked_survival_pct, Q_embarked_survival_pct, C_embarked_survival_pct], label='Survived')
ax.bar(['S', 'Q', 'C'], [S_embarked_unsurvival_pct, Q_embarked_unsurvival_pct, C_embarked_unsurvival_pct], bottom=[S_embarked_survival_pct, Q_embarked_survival_pct, C_embarked_survival_pct], label='Unsurvived')
ax.set_xlabel('Embarked')
ax.set_ylabel('Percentage')
ax.set_title('Percentage of Survived and Unsurvived Passengers by Embarked')
ax.legend()
plt.show()
```
